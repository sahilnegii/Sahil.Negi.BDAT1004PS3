    Occupations
    Step 1. Import the necessary libraries
    Step 2. Import the dataset from this address.
    Step 3. Assign it to a variable called users
    Step 4. Discover what is the mean age per occupation
    Step 5. Discover the Male ratio per occupation and sort it from the most to the least
    Step 6. For each occupation, calculate the minimum and maximum ages
    Step 7. For each combination of occupation and sex, calculate the mean age
    Step 8. For each occupation present the percentage of women and men

```python
# Step 1: Importing the necessary libraries
import pandas as pd
import numpy as np
```

```python
# Step 2 & 3: Importing the dataset from the address given & assigning it to a variable called users
users = pd.read_csv(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep = '|')

# Displaying the first 10 records
users.head(10)
```

```python
# Step 4: Discover what is the mean age per occupation
mean_age = users.groupby('occupation')['age'].mean()

# Displaying the mean age per occupation
mean_age
```

```python
# Step 5: Discover the Male ratio per occupation and sort it from the most to the least
# Calculate the male ratio by grouping data by 'occupation' and applying a lambda function
# Lambda function computes the ratio of males ('M') to the total number of individuals in each occupation
male_ratio = users.groupby('occupation').apply(lambda x: sum(x['gender'] == 'M') / len(x))

# Sort the male_ratio Series in descending order (from most to least)
male_ratio.sort_values(ascending=False, inplace=True)

# Displaying the ratio of males to the total number of individuals for each occupation in the dataset.
male_ratio
```

```python
# Step 6: For each occupation, calculate the minimum and maximum ages
# Group the data by 'occupation' and use the agg() function to calculate both minimum and maximum ages
age_range = users.groupby('occupation')['age'].agg(['min', 'max']).add_suffix('(Age)')

# Displaying the minimum and maximum ages for each occupation in the dataset.
age_range
```

```python
# Step 7: For each combination of occupation and sex, calculate the mean age
# Group the data by both 'occupation' and 'gender' and use the agg() function to calculate the mean age
mean_age = users.groupby(['occupation', 'gender']).agg({'age': 'mean'})

# Displaying the mean age for each combination of occupation and gender in the dataset.
mean_age
```

```python
# Step 8: For each occupation present the percentage of women and men
# Group the data by both 'occupation' and 'gender' and calculate the count of user_ids
num = users.groupby(['occupation', 'gender'])['user_id'].count().rename("count")

# Calculate the percentage by dividing the count by the sum of counts within each occupation group and multiplying by 100
percentage_gender_by_occupation = num / num.groupby(level=0).sum() * 100

percentage_gender_by_occupation
```

```python
# Step 1. Import the necessary libraries
import pandas as pd
import numpy as np
import datetime
```

```python
# Step 2 & step 3 
data = pd.read_csv("C:/Users/sahil/Downloads/Documents/Python Notebooks/wind.txt", sep = "\s+", parse_dates = [[0,1,2]])
data.head()
```

```python
# Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it
def century(x):
    if x.year > 1989:
        year = x.year - 100 
    else:
        year = x.year
    return datetime.date(year, x.month, x.day)


data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(century)
data.head() 
```

```python
# Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])
data = data.set_index('Yr_Mo_Dy')
data.head()
```

```python
# Step 6. Compute how many values are missing for each location over the entire record. They should be ignored in all calculations below.
data.isnull().sum()
```

```python
# Step 7. Compute how many non-missing values there are in total
data.notnull().sum()
```

```python
print (f'Total number of non-missing values is:', data.notnull().sum().sum())
```

```python
# Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the time
mean_wind_speed = data.mean().mean()
mean_wind_speed
```

```python
# Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days
loc_stats = pd.DataFrame({
    'min': data.min(),
    'max': data.max(),
    'mean': data.mean(),
    'std': data.std()
})
loc_stats
```

```python
# Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day. A different set of numbers for each day.
statistics = pd.DataFrame()

statistics['min'] = data.min(axis = 1)
statistics['max'] = data.max(axis = 1)  
statistics['mean'] = data.mean(axis = 1)
statistics['std'] = data.std(axis = 1) 

statistics.head()
```

```python
# Step 11: Find the average windspeed in January for each location. Treat January 1961 and January 1962 both as January.

data.loc[data.index.month == 1].mean()
```

```python
# Step 12. Downsample the record to a yearly frequency for each location.
yearly_data = data.resample('Y').mean()
yearly_data
```

```python
# Step 13. Downsample the record to a monthly frequency for each location.
monthly_data = data.resample('M').mean()
monthly_data
```

```python
# Step 14. Downsample the record to a weekly frequency for each location
weekly_data = data.resample('W').mean()
weekly_data
```

```python
# Step 15: Calculate the min, max, mean, and standard deviation of windspeeds for the first 52 weeks
weekly_stats = data['1961-01-02':'1961-12-31'].resample('W').agg({
    'RPT': ['min', 'max', 'mean', 'std'],
    'VAL': ['min', 'max', 'mean', 'std'],
    'ROS': ['min', 'max', 'mean', 'std'],
    'KIL': ['min', 'max', 'mean', 'std'],
    'SHA': ['min', 'max', 'mean', 'std'],
    'BIR': ['min', 'max', 'mean', 'std'],
    'DUB': ['min', 'max', 'mean', 'std'],
    'CLA': ['min', 'max', 'mean', 'std'],
    'MUL': ['min', 'max', 'mean', 'std'],
    'CLO': ['min', 'max', 'mean', 'std'],
    'BEL': ['min', 'max', 'mean', 'std'],
    'MAL': ['min', 'max', 'mean', 'std']
})

weekly_stats
```
